```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combined_data = pd.merge(city_data, ride_data)

# Display the data table for preview
combined_data
```

```python
# Obtain the x and y coordinates for each of the three city types
urban = combined_data[combined_data["type"] == "Urban"]
suburban = combined_data[combined_data["type"] == "Suburban"]
rural = combined_data[combined_data["type"] == "Rural"]

urban_ride = urban.groupby(["city"]).count()["ride_id"]
urban_fare = urban.groupby(["city"]).mean()["fare"]
urban_count = urban.groupby(["city"]).mean()["driver_count"]

suburban_ride = suburban.groupby(["city"]).count()["ride_id"]
suburban_fare = suburban.groupby(["city"]).mean()["fare"]
suburban_count = suburban.groupby(["city"]).mean()["driver_count"]

rural_ride = rural.groupby(["city"]).count()["ride_id"]
rural_fare = rural.groupby(["city"]).mean()["fare"]
rural_count = rural.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for each city types
plt.scatter(urban_ride, 
            urban_fare, 
            s=10*urban_count, c="lightcoral", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Urban")

plt.scatter(suburban_ride, 
            suburban_fare, 
            s=10*suburban_count, c="lightblue", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Suburban")

plt.scatter(rural_ride, 
            rural_fare, 
            s=10*rural_count, c="gold", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid()

# Create a legend
plt.legend(loc = "upper right")
# Incorporate a text label regarding circle size

plt.text(40, 50, "Note:Circle size correlates with driver count per city.")
# Save Figure
# Show plot
plt.show()
```

```python
# Calculate Type Percents
fare_pie = combined_data.groupby(["type"])["fare"].sum()

# Build Pie Chart
plt.pie(fare_pie, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
# Save Figure
```

```python
# Calculate Ride Percents
ride_pie = 100 * combined_data.groupby(["type"]).count()["ride_id"] / combined_data["ride_id"].count()

# Build Pie Chart
plt.pie(ride_pie, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure

# Show Figure
plt.show()
```

```python
# Calculate City Percents
city_pie = combined_data.groupby(["type"])["driver_count"].sum()

# Build Pie Chart
plt.pie(city_pie, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Save Figure

# Show Figure
plt.show()
```
